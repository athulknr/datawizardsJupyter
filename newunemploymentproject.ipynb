```python
file_path = 'Unemployment_kerala.xlsx'  
data = pd.read_excel(file_path)


total_count = len(data)


branch_counts = data['Month'].value_counts()


plt.figure(figsize=(8, 6))
branch_counts.plot(kind='bar')
plt.title('Month')
plt.xlabel('Month')
plt.ylabel('Estimated Employed')
plt.show()


```
